#IT Incident Intelligence Dashboard & Prediction System

##Exploratory Data Analysis

```python
import pandas as pd
df=pd.read_excel("AnonymizedTickets1.xlsx")
```

```python
df.head()
```

```python
df.info()
```

```python
df.columns = df.columns.str.strip().str.lower().str.replace(" ", "_")
```

```python
df.info()
```

```python
df["reported_date"] = pd.to_datetime(df["reported_date"], errors="coerce")
```

```python
df["resolved_date"] = pd.to_datetime(df["resolved_date"], errors="coerce")
```

```python
df.dtypes
```

```python
df["resolution_time(hours)"].mean()
```

1. Which team that gets most tickets?

```python
team_counts = df["support_group_name"].value_counts()
print(team_counts)
```

2. Which service has the most tickets?

```python
service_counts = df["service"].value_counts()
print(service_counts)
```

3. Which service has the most incidents?

```python
incidents_counts = df[df["incident_type"] == "Incident"]["service"].value_counts()
print(incidents_counts)
```

4. Which service has the most service requests?

```python
SR_counts = df[df["incident_type"] == "Service Request"]["service"].value_counts()
print(SR_counts)
```

```python
category_counts = df.groupby("incident_type")["service"].value_counts()
print(category_counts)
```

```python
pivot = df.pivot_table(
    index="service",
    columns="incident_type",
    aggfunc="size",
    fill_value=0
)

print(pivot)
```

4. SLA non-breach count

```python
sla_non_breach = df[df["slm_real_time_status"] == "Within the Service Target"]["country"].value_counts()
print(sla_non_breach)
```

5. SLA breached count

```python
country_breached = df[df["slm_real_time_status"] == "Service Targets Breached"]["country"].value_counts()
print(country_breached)
```

```python
df["slm_real_time_status"].value_counts()
```

6. Pivot table for country vs status

```python
country_pivot = df.pivot_table(index="country",columns="status",aggfunc="size",fill_value=0)
print(country_pivot)
```
