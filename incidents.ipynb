##Exploratory Data Analysis

```python
import pandas as pd
df=pd.read_excel("AnonymizedTickets1.xlsx")
```

```python
df.head()
```

```python
df.info()
```

```python
df.columns = df.columns.str.strip().str.lower().str.replace(" ", "_")
```

```python
df.info()
```

```python
df["reported_date"] = pd.to_datetime(df["reported_date"], errors="coerce")
```

```python
df["resolved_date"] = pd.to_datetime(df["resolved_date"], errors="coerce")
```

```python
df.dtypes
```

```python
df["resolution_time(hours)"].mean()
```

1. Which team that gets most tickets?

```python
team_counts = df["support_group_name"].value_counts()
print(team_counts)
```

2. Which service has the most tickets?

```python
service_counts = df["service"].value_counts()
print(service_counts)
```

3. Which service has the most incidents?

```python
incidents_counts = df[df["incident_type"] == "Incident"]["service"].value_counts()
print(incidents_counts)
```

4. Which service has the most service requests?

```python
SR_counts = df[df["incident_type"] == "Service Request"]["service"].value_counts()
print(SR_counts)
```

```python
category_counts = df.groupby("incident_type")["service"].value_counts()
print(category_counts)
```

```python
pivot = df.pivot_table(
    index="service",
    columns="incident_type",
    aggfunc="size",
    fill_value=0
)

print(pivot)
```

4. SLA non-breach count

```python
sla_non_breach = df[df["slm_real_time_status"] == "Within the Service Target"]["country"].value_counts()
print(sla_non_breach)
```

5. SLA breached count

```python
country_breached = df[df["slm_real_time_status"] == "Service Targets Breached"]["country"].value_counts()
print(country_breached)
```

```python
df["slm_real_time_status"].value_counts()
```

6. Pivot table for country vs status

```python
country_pivot = df.pivot_table(index="country",columns="status",aggfunc="size",fill_value=0)
print(country_pivot)
```

```python
%config SqlMagic.style = '_DEPRECATED_DEFAULT'
%config SqlMagic.displaylimit = 10
```

```python
#Show first 10 rows of the data
%sql SELECT * FROM incidents LIMIT 10;
```

```python
# How many total tickets are there?
%sql SELECT COUNT(*) FROM incidents;
```

```python
# Number of incidents
%sql SELECT COUNT(*) FROM incidents WHERE incident_type = "Incident";
```

```python
# Show the number of unique services
%sql SELECT DISTINCT service FROM incidents;
```

```python
# Number of service requests
%sql SELECT COUNT(*) FROM incidents WHERE incident_type="Service Request"
```

```python
# Number of tickets per service
%sql SELECT incident_type, COUNT(*) FROM incidents GROUP BY incident_type
```

```python
# Number of tickets per SLA status
%sql SELECT slm_real_time_status, COUNT(*) FROM incidents GROUP BY slm_real_time_status
```

```python
# Services with top 5 number of tickets
%sql SELECT service, COUNT(*) as total FROM incidents GROUP BY "service" ORDER BY total DESC LIMIT 5;
```

```python
# Find average resolution time
%sql SELECT AVG("resolution_time(hours)") FROM incidents
```

```python
# Find average resolution time per service
%sql SELECT service, AVG("resolution_time(hours)") as average FROM incidents GROUP BY service ORDER BY average DESC
```

```python
# Number of tickets created per day and order by number
%sql SELECT DATE(reported_date) AS Date, COUNT(*) AS Count FROM incidents GROUP BY Date ORDER BY Count DESC
```

```python
# Number of tickets created per month
%sql SELECT strftime('%Y-%m', reported_date) as month, COUNT(*) FROM incidents GROUP BY month;
```

```python
# Average resolution time per month
%sql SELECT strftime('%Y-%m', reported_date) as Month, AVG("resolution_time(hours)") AS Average FROM incidents GROUP BY Month
```

```python
# Service that takes the longest to resolve on average
%sql SELECT service, AVG("resolution_time(hours)") AS Average FROM incidents GROUP BY service ORDER BY Average DESC LIMIT 1
```

```python
# Category that is resolved faster
%sql SELECT priority, AVG("resolution_time(hours)") AS Average FROM incidents GROUP BY priority ORDER BY Average
```

```python
# Service that has most tickets
%sql SELECT service, COUNT("Incident_ID") AS count FROM incidents GROUP BY service ORDER BY count DESC
```

```python
# How many tickets were resolved within 24 hours
%sql SELECT COUNT(*) FROM incidents WHERE "resolution_time(hours)" <=24
```

```python
# What is the SLA breach rate
%sql SELECT COUNT(incident_id) AS count, slm_real_time_status FROM incidents GROUP BY slm_real_time_status
```

```python
# Top 3 busiest days by ticket volume
%sql SELECT DATE(reported_date) AS day, COUNT(*) AS count FROM incidents GROUP BY day ORDER BY count DESC LIMIT 3
```

```python
# Service that has the highest SLA breaches
%sql SELECT service, COUNT(*) AS count FROM incidents WHERE slm_real_time_status = "Service Targets Breached" GROUP BY service ORDER BY count DESC
```

```python
# Rank countries by open ticket volume
%sql SELECT country, COUNT(*) AS count FROM incidents WHERE status <> "Resolved" GROUP BY country ORDER BY count DESC
```
